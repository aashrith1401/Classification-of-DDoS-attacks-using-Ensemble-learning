```python
import pandas as pd
```

```python
df=pd.read_csv("D:\\6th Sem\\Mini Project\\cleaned_final_dataset.csv")
```

```python
df.info()
```

```python
y=df.iloc[:,-1]
y = pd.DataFrame(y)
```

```python
print('y\n',y.info())
```

```python
normalized_df=(df-df.mean())/df.std()
normalized_df=normalized_df.drop(' Flow Packets/s',axis=1)
normalized_df=normalized_df.drop('Flow Bytes/s',axis=1)
```

```python
normalized_df.head()
```

```python
normalized_df=normalized_df.drop('Unnamed: 0', axis=1)
normalized_df=normalized_df.drop('Unnamed: 0.2', axis=1) 
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(normalized_df)
```

```python
cluster_labels = kmeans.labels_
```

```python
normalized_df['Cluster'] = cluster_labels
```

```python
normalized_df.describe()
```

```python
normalized_x=normalized_df.iloc[:,normalized_df.columns != 'Label']
```

```python
normalized_x.describe()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, y_train, y_test = train_test_split(normalized_x, y, test_size=0.25, random_state=0)
y_train = np.array(y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
y_train= y_train.ravel()
```

```python
rf_classifier = RandomForestClassifier(n_estimators=500)
knn_classifier = KNeighborsClassifier(n_neighbors=5)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
predictions = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy with RF Percentage:", accuracy*100)
```

```python
knn_classifier.fit(X_train, y_train)
```

```python
predictions1 = knn_classifier.predict(X_test)
```

```python
accuracy1 = accuracy_score(y_test, predictions1)
print("Accuracy with KNN Percentage:", accuracy1*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ensemble_classifier = AdaBoostClassifier(n_estimators=500, learning_rate=0.05)
```

```python
ensemble_classifier.estimators=[knn_classifier, rf_classifier]
```

```python
ensemble_classifier.fit(X_train, y_train)
```

```python
accuracy = ensemble_classifier.score(X_test, y_test)
```

```python
print("Ensemble accuracy percentage:",accuracy*100)
```
