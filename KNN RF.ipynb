```python
x=df.iloc[:,df.columns != 'Label']
y=df.iloc[:,-1]
print("x\n",x.info())
y = pd.DataFrame(y)
print('y\n',y.info())
```

```python
normalized_df=(df-df.mean())/df.std()
normalized_df=normalized_df.drop(' Flow Packets/s',axis=1)
normalized_df=normalized_df.drop('Flow Bytes/s',axis=1)
```

```python
normalized_df.head()
```

```python
normalized_df=normalized_df.drop('Unnamed: 0', axis=1)
normalized_df=normalized_df.drop('Unnamed: 0.2', axis=1) 
```

```python
y=pd.DataFrame(y)
```

```python
normalized_df.describe()
```

```python
normalized_x=normalized_df.iloc[:,normalized_df.columns != 'Label']
```

```python
normalized_x.describe()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, y_train, y_test = train_test_split(normalized_x, y, test_size=0.25, random_state=0)
y_train = np.array(y_train)
```

```python
normalized_x.info()
```

```python
y.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.svm import SVC
```

```python
y_train= y_train.ravel()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=45)
```

```python
rf.fit(X_train, y_train)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions1=knn.predict(X_test)
```

```python
predictions = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy with RF Percentage:", accuracy*100)
```

```python
accuracy1 = accuracy_score(y_test, predictions1)
print("Accuracy with KNN Percentage:", accuracy1*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ensemble_classifier = AdaBoostClassifier(n_estimators=1000, learning_rate=0.1)
```

```python
ensemble_classifier.estimators=[knn, rf]
```

```python
ensemble_classifier.fit(X_train, y_train)
```

```python
accuracy = ensemble_classifier.score(X_test, y_test)
```

```python
print("Ensemble accuracy:",accuracy)
```

```python
print("Ensemble accuracy percentage:",accuracy*100)
```
